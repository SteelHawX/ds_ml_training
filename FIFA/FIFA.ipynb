```python
df_pos_lw = df_position[df_position['Position'] == 'LW']
df_pos_lw['max'] = df_pos_lw.max(axis=1)
df_pos_lw.head()
```

```python
position_labels = position_labels[:-2]
```

```python
for label in position_labels:
    df_pos_lw[label] = df_pos_lw['max'] - df_pos_lw[label]
df_pos_lw.head()
```

```python
se_lw_diff = dict()
for label in position_labels:
    se_lw_diff[label] = df_pos_lw[[label]].mean(axis=0, numeric_only=True)[0]
se_lw_diff['Pos'] = 'LW'
se_lw_diff
```

```python
ext_position_labels = position_labels.copy()
ext_position_labels.append('Pos')
```

```python
df_pos_diff = pd.DataFrame([se_lw_diff])
df_pos_diff.set_index('Pos')
```

```python
df_pos_diff_final = pd.DataFrame()
dict_pos_diff = dict()
for position in position_labels:
    df_pos = df_position[df_position['Position'] == position]
    df_pos['max'] = df_pos.max(axis=1)
    for label in position_labels:
        df_pos[label] = df_pos['max'] - df_pos[label]
    arr_pos_diff = []
    for label2 in position_labels:
        arr_pos_diff.append(df_pos[[label2]].mean(axis=0, numeric_only=True)[0])
    dict_pos_diff[position] = arr_pos_diff
#         dict_pos_diff['Pos'] = position
#         if df_pos_diff_final.empty:
#             df_pos_diff_final = pd.DataFrame([dict_pos_diff])
#             df_pos_diff_final.set_index('Pos')
#         else:
#             df_pos_diff_final.append([dict_pos_diff])
```
